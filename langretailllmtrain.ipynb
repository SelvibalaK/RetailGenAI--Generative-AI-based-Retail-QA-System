```python
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain.utilities import SQLDatabase
from langchain_experimental.sql import SQLDatabaseChain
from langchain.prompts import PromptTemplate

llm = ChatGoogleGenerativeAI(
    model="gemini-pro",
    google_api_key=api_key,
    temperature=0
)

# Database connection
db = SQLDatabase.from_uri("mysql+pymysql://root:root@localhost/atliq_tshirts")
```

```python
# Define a more strict prompt template
PROMPT = """
Based on the database schema below:
{table_info}

Generate a SQL query to answer this question:
{input}

Important: Return ONLY the raw SQL query without any markdown formatting, backticks, or sql tags.
Example format:
SELECT column FROM table WHERE condition;
"""

prompt = PromptTemplate(
    input_variables=["input", "table_info"],
    template=PROMPT
)

# Create the chain with output parsing
db_chain = SQLDatabaseChain.from_llm(
    llm=llm,
    db=db,
    prompt=prompt,
    verbose=True,
    use_query_checker=False,
    return_direct=True
)

# Execute the query
qn1 = db_chain.invoke("How many t-shirts do we have left for levi in large size and white color?")
```

```python
from langchain_community.vectorstores import Chroma
import os

# Create a persistent directory
persist_directory = "chroma_db"
if not os.path.exists(persist_directory):
    os.makedirs(persist_directory)

# Initialize Chroma with persistence
vectorstore = Chroma(
    persist_directory=persist_directory,
    embedding_function=embeddings
)

# Add documents one at a time with explicit persistence
for i in range(len(to_vectorize)):
    vectorstore.add_texts(
        texts=[to_vectorize[i]],
        metadatas=[{'text': str(few_shots[i])}],
    )
    vectorstore.persist()
    
print("Vectorstore creation completed successfully!")
```

```python
# Print current embedding configuration
print(f"Model name: {embeddings.model_name}")
print(f"Encode kwargs: {embeddings.encode_kwargs}")

# Optionally set lighter configuration
embeddings.encode_kwargs['batch_size'] = 1
embeddings.encode_kwargs['show_progress_bar'] = True
```

```python
# Clear any existing encode_kwargs
embeddings.encode_kwargs = {}

# Create the example selector with the corrected configuration
from langchain.prompts import SemanticSimilarityExampleSelector
example_selector = SemanticSimilarityExampleSelector(
    vectorstore=vectorstore,
    k=2,
)

# Run the query and print results
selected_examples = example_selector.select_examples({"Question": "How many Adidas T shirts I have left in my store?"})
print("Selected Examples:")
for i, example in enumerate(selected_examples, 1):
    print(f"\nExample {i}:")
    print(example)
```

```python
from langchain.prompts import FewShotPromptTemplate, PromptTemplate
from langchain.prompts.example_selector.base import BaseExampleSelector

# Define the selector class
class FewShotExampleSelector(BaseExampleSelector):
    def __init__(self, examples, k=3):
        self.examples = examples
        self.k = k
    
    def add_example(self, example):
        self.examples.append(example)

    def select_examples(self, input_variables):
        return self.examples[:self.k]

# Now create the selector instance
example_selector = FewShotExampleSelector(
    examples=few_shots,
    k=3
)

# Create the example prompt template
example_prompt = PromptTemplate(
    input_variables=["Question", "SQLQuery", "SQLResult"],
    template="Question: {Question}\nSQLQuery: {SQLQuery}\nSQLResult: {SQLResult}"
)


# Create the few-shot prompt
few_shot_prompt = FewShotPromptTemplate(
    example_selector=example_selector,
    example_prompt=example_prompt,
    prefix=_mysql_prompt,
    suffix=PROMPT_SUFFIX,
    input_variables=["input", "table_info"]
)

# Create the chain
new_chain = SQLDatabaseChain.from_llm(
    llm=llm,
    db=db,
    prompt=few_shot_prompt,
    verbose=True
)

# Execute the query
result = new_chain.run("How many white color Levi's shirt I have?")
```

```python
import pymysql

# Establish connection to your MySQL database
connection = pymysql.connect(
    host='localhost',  # e.g. 'localhost' or an IP address
    user='root',  # Your MySQL username
    password='root',  # Your MySQL password
    database='atliq_tshirts'  # Your MySQL database name
)

# Create a cursor object
cursor = connection.cursor()

# Define your query
query = """
SELECT `stock_quantity`
FROM `t_shirts`
WHERE `brand` = %s AND `color` = %s AND `size` = %s
LIMIT 5
"""
params = ('Levi', 'White', 'L')

# Execute the query with parameters
cursor.execute(query, params)

# Fetch the results
result = cursor.fetchall()

# Print the results
for row in result:
    print(row)

# Close the cursor and connection
cursor.close()
connection.close()
```
